# Central tendency
**Central tendency** refers to the concept in statistics that identifies the center point or typical value within a data set. It describes where most of the values in a data distribution cluster and is used to provide a single value that best represents the data set. Central tendency is often referred to as the "average" or "middle" of a dataset, but it can be calculated in several different ways, depending on the characteristics of the data.

The three most common measures of central tendency are:

### 1. **Mean (Arithmetic Mean)**

- **Definition**: The mean is the sum of all data points divided by the number of data points.
- **Calculation**: 
  \[
  \text{Mean} = \frac{\sum_{i=1}^{n} x_i}{n}
  \]
  where \( x_i \) represents each data point, and \( n \) is the number of data points.
- **When to use**: The mean is the most commonly used measure of central tendency, especially when the data is symmetrically distributed without extreme outliers.
- **Example**: In the data set \( [2, 4, 6, 8, 10] \), the mean is:
  \[
  \frac{2 + 4 + 6 + 8 + 10}{5} = 6
  \]

### 2. **Median**

- **Definition**: The median is the middle value when all data points are arranged in ascending or descending order. If the data set has an odd number of observations, the median is the middle value. If the data set has an even number of observations, the median is the average of the two middle values.
- **When to use**: The median is particularly useful when dealing with skewed data or data with outliers because it is not affected by extreme values.
- **Example**: For the data set \( [2, 4, 6, 8, 10] \), the median is 6, which is the middle number. For the data set \( [2, 4, 6, 8] \), the median is:
  \[
  \frac{4 + 6}{2} = 5
  \]

### 3. **Mode**

- **Definition**: The mode is the value that appears most frequently in a data set. A data set can have more than one mode (bi-modal or multi-modal) or no mode at all if no number repeats.
- **When to use**: The mode is useful when identifying the most common value, particularly in categorical data.
- **Example**: In the data set \( [2, 3, 3, 4, 5] \), the mode is 3, as it appears most frequently. In the data set \( [1, 1, 2, 3, 3, 4] \), both 1 and 3 are modes (bi-modal distribution).

### Comparison of Measures:

- **Mean**: Sensitive to outliers; good for symmetric distributions.
- **Median**: More robust against outliers; useful for skewed distributions.
- **Mode**: Useful for categorical data or understanding the most frequent observation.

### Example of the Three Measures:
For the data set \( [1, 2, 2, 3, 4, 7] \):
- **Mean**: \( \frac{1 + 2 + 2 + 3 + 4 + 7}{6} = 3.17 \)
- **Median**: The middle two numbers are 2 and 3, so the median is \( \frac{2 + 3}{2} = 2.5 \).
- **Mode**: The most frequent value is 2.

In summary, central tendency measures give insight into the general behavior of a data set by providing a single representative value around which data points tend to cluster. The choice of which measure to use depends on the nature of the data and the specific goals of the analysis.

```python
# Calculate mean, median, and mode
mean = np.mean(data)
median = np.median(data)
mode = stats.mode(data, keepdims=True)[0][0]
```

```python
mean, median, mode
```

```python
# Create the plot
plt.figure(figsize=(10, 6))

# Plot the bell curve (normal distribution)
sns.histplot(data, kde=True, color="skyblue", stat="density", bins=30, label="Data Distribution", alpha=0.5)

# Plot the mean, median, and mode
plt.axvline(mean, color='red', linestyle='--', linewidth=2, label=f'Mean: {mean:.2f}')
plt.axvline(median, color='green', linestyle='-', linewidth=2, label=f'Median: {median:.2f}')
plt.axvline(mode, color='blue', linestyle='-', linewidth=2, label=f'Mode: {mode:.2f}')

# Add labels and title
plt.title('Mean, Median, and Mode on Normal Distribution', fontsize=14)
plt.xlabel('Value', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

```python
import numpy as np

# Define data points and corresponding weights
values = np.array([85, 90, 78])
weights = np.array([3, 2, 5])

# Calculate the weighted mean
weighted_mean = np.average(values, weights=weights)

# Output the result
print(f'The weighted arithmetic mean is: {weighted_mean}')
```

```python
from scipy.stats import trim_mean
import numpy as np

# Define a data set with some outliers
data_with_outliers = np.array([10, 12, 14, 15, 16, 18, 20, 100, 105])

# Calculate the 10% trimmed mean (removes the lowest and highest 10% of values)
trimmed_mean_10 = trim_mean(data_with_outliers, proportiontocut=0.1)

# Output the result
print(f'The 10% trimmed mean is: {trimmed_mean_10}')
```

## Grouped Median

```python
import numpy as np

# Data from the table: Age intervals and frequencies
age_intervals = [(1, 5), (6, 15), (16, 20), (21, 50), (51, 80), (81, 110)]
frequencies = [200, 450, 300, 1500, 700, 44]

# Total number of observations
n = np.sum(frequencies)

# Calculate cumulative frequencies
cumulative_frequencies = np.cumsum(frequencies)

# Median class is the class where the cumulative frequency just crosses n/2
median_position = n / 2

# Finding the median class index
median_class_index = np.where(cumulative_frequencies >= median_position)[0][0]

# Extracting the details for the median class
L1 = age_intervals[median_class_index][0]  # Lower boundary of the median class
freq_median = frequencies[median_class_index]  # Frequency of the median class
cumulative_freq_before = cumulative_frequencies[median_class_index - 1] if median_class_index > 0 else 0
class_width = age_intervals[median_class_index][1] - age_intervals[median_class_index][0]  # Class width

# Applying the formula
median = L1 + ((n / 2 - cumulative_freq_before) / freq_median) * class_width

print(f'The approximate median is: {median}')
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Generate data for a normal distribution
mean = 0
std_dev = 1
x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x, mean, std_dev)

# Create the plot
plt.figure(figsize=(10, 6))

# Plot the normal distribution curve
plt.plot(x, y, label='Normal Distribution', color='blue')

# Mark mean, 1st, 2nd, and 3rd standard deviations
plt.axvline(mean, color='red', linestyle='--', label='Mean (0)')
plt.axvline(mean - std_dev, color='green', linestyle='--', label='1 SD (-1, 1)')
plt.axvline(mean + std_dev, color='green', linestyle='--')
plt.axvline(mean - 2*std_dev, color='orange', linestyle='--', label='2 SD (-2, 2)')
plt.axvline(mean + 2*std_dev, color='orange', linestyle='--')
plt.axvline(mean - 3*std_dev, color='purple', linestyle='--', label='3 SD (-3, 3)')
plt.axvline(mean + 3*std_dev, color='purple', linestyle='--')

# Fill the areas under the curve
plt.fill_between(x, y, where=(x >= mean - std_dev) & (x <= mean + std_dev), color='green', alpha=0.2, label='68% within 1 SD')
plt.fill_between(x, y, where=(x >= mean - 2*std_dev) & (x <= mean + 2*std_dev), color='orange', alpha=0.2, label='95% within 2 SD')
plt.fill_between(x, y, where=(x >= mean - 3*std_dev) & (x <= mean + 3*std_dev), color='purple', alpha=0.2, label='99.7% within 3 SD')

# Add labels and title
plt.title('Normal Distribution with Standard Deviations', fontsize=14)
plt.xlabel('Value', fontsize=12)
plt.ylabel('Probability Density', fontsize=12)
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```
