```python
import numpy as np

# Example dataset
data = np.array([10, 20, 30, 40, 50])

# Apply Min-Max Normalization
min_val = np.min(data)
max_val = np.max(data)

# Perform the normalization for each value in the data
normalized_data = (data - min_val) / (max_val - min_val)

print("Original Data:", data)
print("Min-Max Normalized Data:", normalized_data)
```

```python
import numpy as np

# Example data
data = np.array([73_600, 54_000, 45_500, 80_000])

# Given mean and standard deviation (you can compute these from the dataset if not provided)
mu = np.mean(data)       # Mean
sigma = np.std(data)     # Standard deviation

# Z-score normalization formula: v' = (v - mu) / sigma
z_score_normalized = (data - mu) / sigma

print("Original Data:", data)
print("Z-score Normalized Data:", z_score_normalized)
```

```python
import numpy as np

# Example data
data = np.array([73_600, 54_000, 45_500, 80_000])

# Find the maximum absolute value
max_val = np.max(np.abs(data))

# Determine j such that max_val / (10^j) < 1
j = np.ceil(np.log10(max_val))

# Decimal scaling normalization formula: v' = v / 10^j
decimal_scaling_normalized = data / (10**j)

print("\nDecimal Scaling Normalized Data:", decimal_scaling_normalized)
```
